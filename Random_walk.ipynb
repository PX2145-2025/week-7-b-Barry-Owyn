# Introduction
Random processes are very common in physics. When dealing with large numbers of molecules, such as in gases, we donâ€™t describe the trajectory of every single particle ($10^{23}$ particles is too much to handle) but often use probabilities to determine the average behaviour. Here, we look at some illustrative examples of using random numbers in physical applications.

## Random walks
The random walk problem is a classic example which, despite its simplicity, describes very well the basic physical phenomenon of _diffusion_. Diffusion is a process where par- ticles which are initially concentrated in a region in space spread out with time. For example, a drop of ink in water will gradually spread to the whole volume of water, even if the water is perfectly still. This is because the individual molecules (both the ink and the water) (i) have random directions and speeds, which depend on the temperature, and (ii) also collide with each other. It turns out that these two ingredients can be modelled very easily.

## The method
We first look at a random walk in 1-D for a single particle. We start the particle off at the origin of the x-axis and assume that it moves unit distance in each time step (the molecular speed is thus 1 unit). At the end of each time step we set the direction to left or right with equal probability, mimicking the effect of collisions. You can easily write the python code, but for flexibility of later python coding, I show here how to define a function, called `walk(steps)`, which is passed a single number steps and returns a vector of that length, which contains all the positions of the random walk.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    '''
    x = np.zeros(steps, int)  # initialise the array of integers
    x[0] = 0  # start at the origin
    for i in range(1, len(x)):
        rnd = random()  # rnd selected from [0,1)
        # step left or right randomly ...
        if (rnd<0.5):
            x[i] = x[i-1] + 1  # steps right one unit
        else:
            x[i] = x[i-1] - 1 # steps left one unit
    return x
````

IMPORTANT: Make sure you understand how this function is used, and how it works. The rest of this worksheet is based on using it. You can also use the vectorised version below.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    Vectorised method.
    '''
    x = np.random.choice([-1, 1], size=steps) # random -1 or 1
    x = np.cumsum(x) # cumulative sum
    return x
````

Those two quantities, the "mean position" and the "mean squared position" are standard quantities in random walks.
````python
# Question 2

# Create the 2D array which will store both the position of the particle and it's step. Each column represents a single walk, and each row represents a single step.
particles = np.zeros((1000, 10), int)
N = 10
steps = 1000

# Fill the empty array with the steps data for each of the N particles.
for i in range(N):
    particles[:, i] = walk(steps)
    plt.plot(particles[:, i], label=f'Walker {i+1}') # Plots the 10 walkers, with the inputted walk data plotted against the number of steps utilising the 2dD array.
    # Each walker is uniquely labelled using the i interger within the loop, increasing by 1 every time.

plt.xlabel("Step")
plt.ylabel("1D Position")
plt.title(f"{N} 1D Random Walkers ({steps} steps each)")
plt.legend()
````

````python
# Create the 2D array which will store both the position of the particle and it's step. Each column represents a single walk, and each row represents a single step.
particles2 = np.zeros((1000, 500), int)
N = 500
steps = 1000

# Fill the empty array with the steps data for each of the N particles.
for i in range(N):
    particles2[:, i] = walk(steps)
    plt.plot(particles2[:, i]) # Plots the 10 walkers, with the inputted walk data plotted against the number of steps utilising the 2dD array.
    # Each walker is uniquely labelled using the i interger within the loop, increasing by 1 every time.

plt.xlabel("Step")
plt.ylabel("1D Position")
plt.title(f"{N} 1D Random Walkers ({steps} steps each)")

# Define an array of the last step positions, achieved by simply retrieving all of the values in the final row of the particles2 2D array.
last_step_positions = particles2[999]
mean_position = np.mean(last_step_positions)
mean_squared_position = np.mean(last_step_positions**2)

print('Mean position at the last step:', mean_position)
print('Mean squared position at the last step:', mean_squared_position)
````

````python
# Calculate the mean position at each step. Since particles2 is a 2D array, it is simple to find the mean of every single
# row of the 2D array by using axis=1. Since every row of the array is a different step, this is therefore calculating
# the mean at every individual step from the positions of 500 different particles at this step and putting it into an array, which can then subsequently be plotted.
mean_positions = np.mean(particles2, axis=1)
plt.plot(np.linspace(1, 1000, 1000), mean_positions)
plt.xlabel("Step")
plt.ylabel("Mean Position in 500 Walks")
plt.title("Mean position of a particle over 1000 steps of 500 random walks.")
````

````python
# The above process can be repeated for the mean squared positions, simkply changing np.mean to have the values within each row of particles2 squared before being averaged.
mean_squared_positions = np.mean(particles2**2, axis=1)

plt.plot(mean_squared_positions)
plt.xlabel("Step")
plt.ylabel("Mean Squared Position in 500 Walks")
plt.title("Mean squared position of a particle over 1000 steps of 500 random walks.")

# It is interesting to note that the mean squared position of the particle approximately follow the relationship y=x, forming a diagonal line of gradient +1.
# As the step increases, the mean squared position increases proportionally so, almost exactly, whereby e.g. at step 500 the mean squared position is likely to be 500.
````
